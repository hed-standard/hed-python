## Preparation of the Sternberg dataset

This script sets up the event files for the Sternberg dataset
and does basic checks to see agreement between `EEG.event` in
the `.set` file and the corresponding `_events.tsv` file.

The script assumes that the data is in BIDS format and that each BIDS events
file of the form `_events.tsv` has a corresponding events file with
suffix `_events_temp.tsv` that was previously dumped from the `EEG.set` files.

```python
from hed.tools import get_file_list
bids_root_path = 'G:/Sternberg/SternbergWorking/'
event_files_bids = get_file_list(bids_root_path, types=[".tsv"], suffix="_events")
event_files_eeg = get_file_list(bids_root_path, types=[".tsv"], suffix="_events_temp")
bids_skip = ['onset', 'duration', 'sample']
eeg_skip = ['latency', 'urevent', 'ReqTime', 'ReqDur', 'init_index', 'init_time',
            'event_code', 'cond_code', 'sample_offset', 'duration', 'TTime']
```

### Verify same column names

Create a dictionary with keys of form subj#_run# and values which are
full file paths to allow the event files in different forms to be compared.
Print the column names for each file to make sure all the same.

```python
import os
from hed.tools import get_new_dataframe

bids_file_dict = {}
print("\nBIDS form of the events:")
for file in event_files_bids:
    base = os.path.basename(file)
    pieces = base.split('_')
    key = f"{pieces[0]}_{pieces[-2]}"
    df = get_new_dataframe(file)
    bids_file_dict[key] = file
    print(f"{key}: {str(list(df.columns.values))}")

print(f"{len(list(bids_file_dict))} BIDS style event files")
```

```python
eeg_file_dict = {}
print("\nEEG form of the events:")
for file in event_files_eeg:
    base = os.path.basename(file)
    pieces = base.split('_')
    key = f"{pieces[0]}_{pieces[-3]}"
    df = get_new_dataframe(file)
    eeg_file_dict[key] = file
    print(f"{key}: {str(list(df.columns.values))}")
```

## Create summary dictionaries

The next two sections create summary dictionaries

```python
from hed.tools import ColumnDict
bids_dicts_all = ColumnDict(skip_cols=bids_skip, name=f"{bids_root_path} BIDS" )
bids_dicts = {}
for key, file in bids_file_dict.items():
    orig_dict = ColumnDict(skip_cols=bids_skip, name=f"{file} BIDS")
    orig_dict.update(file)
    bids_dicts[key] = orig_dict
    bids_dicts_all.update_dict(orig_dict)
print('BIDS events summary:')
bids_dicts_all.print()
```

```python
from hed.tools import ColumnDict
eeg_dicts_all = ColumnDict(skip_cols=eeg_skip, name=f"{bids_root_path} EEG" )
eeg_dicts = {}
for key, file in eeg_file_dict.items():
    eeg_dict = ColumnDict(skip_cols=eeg_skip, name=f"{file} EEG")
    eeg_dict.update(file)
    eeg_dicts[key] = eeg_dict
    eeg_dicts_all.update_dict(eeg_dict)
# print('EEGLAB events summary:')
# eeg_dicts_all.print()
```

```python
cat = eeg_dicts_all.categorical_info
sorted_keys = sorted(cat.keys())
for key in sorted_keys:
    value_dict = cat[key]
    x = list(value_dict)

    sorted_v_keys = sorted(map(str, x))
    print(f"{4 * 2}{key} ({len(sorted_v_keys)} distinct values):")
    for key1 in x:
        if key1:
            print(f"{key1} : {value_dict[key1]}")
        else:
            print(f"Empty[{key1}] : {value_dict[key1]}")
    # sorted_v_keys = sorted(value_dict.keys())
```

## Recheck of Sternberg after final restructuring

```python
from hed.tools import get_file_list, get_new_dataframe, make_file_dict

file_root = 'G:/Sternberg/SternbergInfo/'
files_bids = get_file_list(file_root, types=[".tsv"], suffix="_events")
files_dict = make_file_dict(files_bids)

print(f"\nChecking the duration")
for key, file in files_dict.items():
    df = get_new_dataframe(file)
    letter_mask = df['event_type'].map(str) == 'show_letter'
    role_mask = (df['task_role'].map(str) == 'nontarget') | (df['task_role'].map(str) == 'target')
    duration_max = df.loc[letter_mask & role_mask, 'duration'].max()
    duration_min = df.loc[letter_mask & role_mask, 'duration'].min()
    print(f"{key}: max letter duration = {duration_max} min letter duration = {duration_min}")
```

## Resample the events file and move up

This script assumes that `_events.tsv` in the `eeg` directory satisfies the following:
1) The sampling rate was 250 Hz.
2) The first event has a sample of 1.0  its onset should be changed from 1/250 to 1/1100.
3) The other events should have sample recomputed as onset*1100.

```python
import os
from hed.tools.io_utils import get_file_list
from hed.tools.data_utils import get_new_dataframe

new_sample_rate = 1100.0
bids_root_path = 'G:/WH_working1'
event_files_bids = get_file_list(bids_root_path, extensions=[".tsv"], name_suffix="_events")
for my_file in event_files_bids:
    print(my_file)
    df = get_new_dataframe(my_file)
    df.loc[df.index[0], 'onset'] = 1./new_sample_rate
    df.loc[1:, 'sample'] = df.loc[1:, 'onset']*new_sample_rate

    components1 = os.path.split(my_file)
    components2 = os.path.split(components1[0])
    filename = os.path.join(components2[0], components1[1])
    print(f"new file: {filename}")
    df.to_csv(filename, sep='\t', index=False, float_format="%.8f")
    os.remove(my_file)
```

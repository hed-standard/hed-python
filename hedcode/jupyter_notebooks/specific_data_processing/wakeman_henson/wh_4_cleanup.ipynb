The final cleanup assumes that we copy the entire BIDS dataset into
new directory and delete in the new directory so that we still have
the original directory to work with.

```python
import os
from hed.tools.io_utils import get_file_list

bids_root_path = 'G:\WH_working3'
file_list = get_file_list(bids_root_path, extensions=[".tsv"], name_suffix="_events")
for file in file_list:
    os.remove(file)

file_list = get_file_list(bids_root_path, extensions=[".tsv"], name_suffix="_temp1")
for file in file_list:
    os.remove(file)

file_list = get_file_list(bids_root_path, extensions=[".tsv"], name_suffix="_temp2")
for file in file_list:
    new_name = file[:-10] + ".tsv"
    os.rename(file, new_name)
    print(new_name)
```

This script removes the extra stim files in a BIDS example dataset. It first makes a hash table of
the stim files listed in the events.tsv files.  It then removes those files in the stimuli
directory that aren't there.

```python
import os
from hed.tools.io_utils import get_file_list
from hed.tools.data_utils import get_new_dataframe

# stim_file_path = 'D:/Research/HED/hed-examples/datasets/eeg_ds003654s/stimuli'
# bids_root_path = 'D:/Research/HED/hed-examples/datasets/eeg_ds003654s'
stim_file_path = 'D:/Research/HED/hed-examples/datasets/eeg_ds003654s_inheritance/stimuli'
bids_root_path = 'D:/Research/HED/hed-examples/datasets/eeg_ds003654s_inheritance'

event_file_list = get_file_list(bids_root_path, extensions=[".tsv"], name_suffix="_events")
stim_dict = {}
for file in event_file_list:
    df = get_new_dataframe(file)
    s_files = df['stim_file'].tolist()
    for s_file in s_files:
        stim_dict[s_file] = ""

stim_file_list = get_file_list(stim_file_path, extensions=[".bmp"])
remove_count = 0
for file in stim_file_list:
    file_name = os.path.basename(file)
    if file_name not in stim_dict:
        os.remove(file)
        print(f"{file_name}: {file}")
        remove_count += 1
print(f"{stim_dict}")
print(f"{remove_count} stim files removed")
```
